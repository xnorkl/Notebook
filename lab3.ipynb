```python
#first derivative

def g(t):
    t = t**4
    return t

def diff (f,x,h=1E-6):
    r = (f(x+h)-f(x-h))/(2*h)
    return r

def diff2 (f,x,h=1E-6):
    r = (f(x-h)-2*(f(x))+f(x+h))/(h**2)
    return r
```
